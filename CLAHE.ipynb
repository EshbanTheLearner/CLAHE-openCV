```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_v1 = cv2.imread("img/test.jpeg")
```

```python
plt.imshow(img_v1)
plt.show()
```

```python
img = cv2.imread("img/test.jpeg", 0)
plt.imshow(img)
plt.show()
```

```python
plt.hist(img.flat, bins=100, range=(0, 255))
plt.show()
```

```python
equ = cv2.equalizeHist(img)
plt.imshow(equ)
plt.show()
```

```python
plt.hist(equ.flat, bins=100, range=(0, 255))
plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
cl_img = clahe.apply(img)
plt.imshow(cl_img)
plt.show()
```

```python
plt.hist(cl_img.flat, bins=100, range=(0, 255))
plt.show()
```

```python
ret, thresh_1 = cv2.threshold(cl_img, 190, 150, cv2.THRESH_BINARY)
ret, thresh_2 = cv2.threshold(cl_img, 190, 255, cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh_1)
plt.show()
```

```python
plt.imshow(thresh_2)
plt.show()
```

```python
ret, thresh_3 = cv2.threshold(cl_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(thresh_3)
plt.show()
```
